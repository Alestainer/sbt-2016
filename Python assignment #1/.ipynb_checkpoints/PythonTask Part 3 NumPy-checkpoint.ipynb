```python
def my_linspace(start, stop, num):
    return [float(i)/num * (stop-start) + start for i in range(num)]
```

```python
print (my_linspace(-1, 1, 50))
```

Matplotlib inline makes graphs appear in the notebook

```python
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

```python
x = np.linspace(-10,10,100)
y = map(lambda p: math.cos(p), x)
z = map(lambda p: math.sin(p), x)
plt.plot(x, y, 'o', label = 'cos')
plt.plot(x, z, label = 'sin')
plt.legend()
plt.show()
```

```python
a = np.linspace(0.1,5,200)
exp = map(lambda x: math.exp(x), a)
quadr = a ** 2
qubic = a ** 3
log = map(lambda x: math.log(x), a)
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(223)
ax1.plot(a, exp)

ax2 = fig.add_subplot(224)
ax2.plot(a, log, 'k-')

ax3 = fig.add_subplot(221)
ax3.plot(a, quadr, 'b-')

ax4 = fig.add_subplot(222)
ax4.plot(a, qubic, 'g-')


plt.tight_layout()
fig = plt.gcf()
```

```python
def fun(x, y):
    return x*3*y + x - 2*y

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
_X, _Y = np.meshgrid(x, y)
zs = np.array([fun(x,y) for x,y in zip(np.ravel(_X), np.ravel(_Y))])
Z = zs.reshape(_X.shape)

ax.plot_surface(_X, _Y, Z, cmap = cm.coolwarm)
```

```python
x = np.linspace(0,1, 100)
y = x * 100
f = np.ndarray(shape = (100, 100))
for i in range(100):
    for j in range(100):
        f[i][j] = x[i] * y[j]

f_min, f_max = f.min(), f.max()
```

```python
plt.pcolormesh(x, y, f, cmap = cm.coolwarm, vmin = f_min, vmax = f_max)
```
